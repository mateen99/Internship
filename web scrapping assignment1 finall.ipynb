# Q2).Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)

```python
page2 = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
soup2 = BeautifulSoup(page2.content)
```

```python
name = soup2.find_all("div",class_="lister-item mode-advanced")
name
```

```python
movies_name=[]
rating2=[]
year=[]
```

```python
for i in name:
    movie=i.h3.find("a")
    movies_name.append(movie.text)
   
```

```python
for i in name:
    rate=i.find("div",class_="inline-block ratings-imdb-rating").text.replace("\n","")
    rating2.append(rate)
```

```python
for i in name:
    yr=i.find("span",class_="lister-item-year text-muted unbold").text.replace("(","").replace(")","")
    year.append(yr)
```

```python
len(movies_name)
```

```python
len(rating2)
```

```python
len(year)
```

```python
imdb_data_1 = pd.DataFrame({"Movie Name":movies_name,"Ratings":rating2,"Year of Release":year})
```

```python
imdb_data_1
```

# Q9) Write a python program to scrape mentioned details from dineout.co.in :i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python
page1 = requests.get("https://www.dineout.co.in/delhi-restaurants/buffet-special")
```

```python
page1
```

```python
soup = BeautifulSoup(page1.content)
```

```python
rest=[]
for i in soup.find_all("a",class_="restnt-name ellipsis"):
    a=i.text
    rest.append(a)
rest
```

```python
cus=[]
for i in soup.find_all("span",class_="double-line-ellipsis"):
    cus.append(i.text.split()[6:9])
cus 
```

```python
loc=[]
for i in soup.find_all("div",class_="restnt-loc ellipsis"):
    a=i.text
    loc.append(a)
loc   
```

```python
rat=[]
for i in soup.find_all("div",class_="restnt-rating rating-4"):
    a=i.text
    rat.append(a)
rat   
```

```python
im=[]
for i in soup.find_all("img",class_="no-img"):
    a=i['data-src']
    im.append(a)
im   
```

```python
data_dine=pd.DataFrame({"Restaurant":rest,"Cuisine":cus,"Location":loc,"Rating":rat,"URL":im})
```

```python
data_dine
```
